```python
import pandas as pd
```

```python
player_matches['player_rank'].fillna(999,inplace=True)
player_matches['opp_rank'].fillna(999,inplace=True)
player_matches['player_age'].fillna(player_matches['player_age'].median(),inplace=True)
player_matches['opp_age'].fillna(player_matches['player_age'].median(),inplace=True)
player_matches['player_ht'].fillna(player_matches.groupby('player_hand')['player_ht'].transform('median'),inplace=True)
player_matches['opp_ht'].fillna(player_matches.groupby('opp_hand')['opp_ht'].transform('median'),inplace=True)
player_matches.dropna(subset=['player_aces','player_df','opp_aces','opp_df'],inplace=True)
player_matches['surface'].fillna(player_matches['surface'].mode()[0],inplace=True)
```

```python
def calculate_elo(df,k_factor=32,intial_elo=1500):
    df=df.sort_values('tourney_date').reset_index(drop=True)

    player_elo={}
    surface_elo={}

    elo_list=[]
    opp_elo_list=[]
    surface_elo_list=[]
    opp_surface_elo_list=[]

    for idx,row in df.iterrows():
        player=row['player']
        opponent=row['opponent']
        surface=row['surface']


        player_current_elo=player_elo.get(player,intial_elo)
        opp_current_elo=player_elo.get(opponent,intial_elo)

        player_surf_elo=surface_elo.get((player,surface),intial_elo)
        opp_surf_elo=surface_elo.get((opponent,surface),intial_elo)

        elo_list.append(player_current_elo)
        opp_elo_list.append(opp_current_elo)
        surface_elo_list.append(player_surf_elo)
        opp_surface_elo_list.append(opp_surf_elo)

        expected_player=1/(1+10**((opp_current_elo-player_current_elo)/400))
        expected_opp=1-expected_player

        actual_player=row['win']
        actual_opp=1-actual_player

        player_elo[player]=player_current_elo+k_factor*(actual_player-expected_player)
        player_elo[opponent]=opp_current_elo+k_factor*(actual_opp-expected_opp)

        expected_surf=1/(1+10**((opp_surf_elo-player_surf_elo)/400))
        surface_elo[(player,surface)]=player_surf_elo+k_factor*(actual_player-expected_surf)
        surface_elo[(opponent,surface)]=opp_surf_elo+k_factor*(actual_opp-(1-expected_surf))

    df['player_elo']=elo_list
    df['opp_elo']=opp_elo_list
    df['player_surface_elo']=surface_elo_list
    df['opp_surface_elo']=opp_surface_elo_list
    df['elo_diff']=df['player_elo'] - df['opp_elo']
    return df
player_matches=calculate_elo(player_matches)
```

```python
def add_rolling(df,window=5):
    df=df.sort_values(['player','tourney_date'])

    #average for player stats
    df['player_average_aces']=df.groupby('player')['player_aces'].transform(lambda x:x.rolling(window,min_periods=1).mean().shift(1))

    df['player_avg_df']=df.groupby('player')['player_df'].transform(lambda x:x.rolling(window,min_periods=1).mean().shift(1))

    df['player_win_rate']=df.groupby('player')['win'].transform(lambda x:x.rolling(window,min_periods=1).mean().shift(1))

    df['player_surface_win_rate']=df.groupby(['player','surface'])['win'].transform(lambda x:x.rolling(window,min_periods=1).mean().shift(1))

    return df
```

```python
player_matches=add_rolling(player_matches,window=5)
```

```python
def add_rolling_opp(df,window=5):
    df=df.sort_values(['opponent','tourney_date'])

    #average for player stats
    df['opp_average_aces']=df.groupby('opponent')['opp_aces'].transform(lambda x:x.rolling(window,min_periods=1).mean().shift(1))

    df['opp_avg_df']=df.groupby('opponent')['opp_df'].transform(lambda x:x.rolling(window,min_periods=1).mean().shift(1))

    df['opp_win_rate']=df.groupby('opponent')['win'].transform(lambda x:(1-x).rolling(window,min_periods=1).mean().shift(1))

    df['opp_surface_win_rate']=df.groupby(['opponent','surface'])['win'].transform(lambda x:(1-x).rolling(window,min_periods=1).mean().shift(1))

    return df
```

```python
player_matches=add_rolling_opp(player_matches,window=5)
```

```python
from sklearn.preprocessing import LabelEncoder
le_surface=LabelEncoder()
player_matches['surface']=le_surface.fit_transform(player_matches['surface'])
le_level=LabelEncoder()
player_matches['tourney_level']=le_level.fit_transform(player_matches['tourney_level'])
le_round=LabelEncoder()
player_matches['round']=le_round.fit_transform(player_matches['round'])
l_hand=LabelEncoder()
player_matches['player_hand']=l_hand.fit_transform(player_matches['player_hand'])
player_matches['opp_hand']=l_hand.fit_transform(player_matches['opp_hand'])
```

```python
features = [
    'surface', 'tourney_level', 'round', 'best_of',
    'player_rank', 'opp_rank', 'player_age', 'opp_age',
    'player_hand', 'opp_hand', 'player_ht', 'opp_ht',
    'player_average_aces', 'player_avg_df', 'player_win_rate', 'player_surface_win_rate',
    'opp_average_aces', 'opp_avg_df', 'opp_win_rate', 'opp_surface_win_rate','player_elo',
       'opp_elo', 'player_surface_elo', 'opp_surface_elo'
]
```

```python
X=player_matches[features]
Y=player_matches['win']
print(Y.shape,X.shape)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train , X_test , Y_train , Y_test= train_test_split(X,Y,test_size=0.2,random_state=10)
```

```python
X_train.shape , Y_train.shape
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score , classification_report
```

```python
xgb=XGBClassifier(n_estimators=200,learning_rate=0.1,max_depth=6)
```

```python
xgb.fit(X_train,Y_train)
```

```python
y_pred=xgb.predict(X_test)
```

```python
print(accuracy_score(Y_test,y_pred))
print(classification_report(Y_test,y_pred))
```

```python
a=pd.DataFrame({
    'feature': features,
    'importance': xgb.feature_importances_
}).sort_values('importance', ascending=False)
print(a)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
plt.bar(a['feature'],a['importance'])
plt.xticks(rotation=45,ha='right')
plt.ylabel("Importance")
plt.xlabel("Features")
plt.show()
```
